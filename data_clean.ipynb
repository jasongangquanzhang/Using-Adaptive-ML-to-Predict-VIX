```python
import pandas as pd
import datetime
from imp import reload

import numpy as np
from dateutil.relativedelta import *
```

# Import Data

```python
data_uri = "calendar.xlsx"
df_data = pd.read_excel(data_uri,header = [0,1],skiprows=[0])
df_data

df_data = df_data.rename(columns={"Unnamed: 17_level_1":'',"Unnamed: 18_level_1":""})
```

```python
df_data.columns = df_data.columns.map(lambda x: '_'.join(map(str, x)))
df_data
```

```python
df_data = df_data.rename(columns = {"Unnamed: 0_level_0_Unnamed: 0_level_1":"datetime"}).set_index("datetime")
df_data
```

```python
# date of business day
trade_day = df_data["SPX Index_VOLUME"].dropna().index.to_list()
trade_day
```

```python
# some data has different frequence
# we use front fill
df_data = df_data.fillna(method="ffill",axis=0)
df_data = df_data.loc[trade_day]
df_data
```

```python
df = df_data.dropna(axis = 1,how='all')
```

# Export data

```python
df.to_csv("cleaned.csv")
```
