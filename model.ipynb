```python
import pandas as pd
import datetime
from imp import reload
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from dateutil.relativedelta import *
```

```python
data_uri = "cleaned.csv"
df = pd.read_csv(data_uri)
df
```

```python
df = df.set_index("datetime")
df
```

# Initial training and testing

```python
df.index = df.index.astype("datetime64[ns]")
df.index
```

```python
train_start = df.index[259] #Timestamp('1994-01-03 00:00:00')
train_end = df.index[259+3999]# Timestamp('2009-11-09 00:00:00')
```

```python
test_start = df.index[4295]# Timestamp('2010-01-04 00:00:00')
test_end =  df.index[4295+2768] # Timestamp('2020-12-31 00:00:00')
```

## Initial Y

```python
y = df[["VIX Index_CHG_PCT_1D"]]
y
```

```python
# we only care about long or short 
y["VIX Index_CHG_PCT_1D"] =y["VIX Index_CHG_PCT_1D"].apply(lambda s: 1  if s>0 else 0)
y
```

```python
# also we want to use VIX_t predict VIX_t+1
y = y.shift(-1)
y
```

```python
y_train = y.loc[train_start:train_end]
y_test = y.loc[test_start:test_end]
```

## Initial X

```python
x = df.drop("VIX Index_CHG_PCT_1D",axis = 1)
x
```

```python
x_train = x[train_start:train_end] # train + 5-fold cv
x_test = x[test_start:test_end] # out of sample
```

## Standardization on X

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
scaler.fit(x_train) # avoid look forward
x_train_standardized = scaler.transform(x_train)
x_test_standardized = scaler.transform(x_test)
```

```python
x_out_v = x_train_standardized[-400:]
y_out_v = y_train[-400:]
x_cv = x_train_standardized[:3600]
y_cv = y_train[:3600]
```

```python
y_cv = y_cv["VIX Index_CHG_PCT_1D"].to_list()
y_out_v = y_out_v["VIX Index_CHG_PCT_1D"].to_list()
```

```python
train_accuracies = []
test_accuracies = []
models = ["NB","LR","DT","RF","AB","MLP","ENS"]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
var_smoothing_range = np.logspace(-20, -8, num=20)
var_smoothing_range
```

```python
NB_param_grid = {
    'var_smoothing': var_smoothing_range
}
```

## Grid search

```python
NB_grid_search = GridSearchCV(GaussianNB(), NB_param_grid, cv=5,n_jobs=-1)
NB_grid_search.fit(x_cv, y_cv)
print("Best hyperparameters:", NB_grid_search.best_params_)
NB_grid_search.best_params_
```

## train model

```python
NB_best_model = NB_grid_search.best_estimator_
NB_best_model.fit(x_cv, y_cv)
```

```python
y_train_pred = NB_best_model.predict(x_cv)
y_test_pred = NB_best_model.predict(x_out_v)
NB_train_accuracy = accuracy_score(y_cv, y_train_pred)
NB_test_accuracy = accuracy_score(y_out_v, y_test_pred)
train_accuracies.append(NB_train_accuracy)
test_accuracies.append(NB_test_accuracy)
print("Train accuracy:", NB_train_accuracy)
print("Test accuracy:", NB_test_accuracy)
```
